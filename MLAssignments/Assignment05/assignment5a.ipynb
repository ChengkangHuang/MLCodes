```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
file = "_DSC0003.jpg"
living_room = io.imread(file)
plt.figure()
plt.imshow(living_room)
plt.show()
```

```python
def draw(N_cluster,data):
    shape=data.shape
    new_data = data.reshape(shape[0]*shape[1], 3)
    clusting = KMeans(n_clusters=N_cluster)
    cls_model = clusting.fit(new_data)
    centroids = cls_model.cluster_centers_
    cls_pred = cls_model.labels_
    for i in range(0, N_cluster):
        new_data[cls_pred == i] = centroids[i]
    new_data2 = new_data.reshape(shape[0], shape[1], 3)
    plt.figure()
    plt.title(file+" - k = "+str(N_cluster))
    plt.imshow(new_data2)
    return cls_model.inertia_

```

```python
inertias=[]
for i in reversed(range(2,20,2)):
    # sse=k_Means_inertia(cat, i)
    sse=draw(i, living_room)
    print("k=%s\t SSE = %f"%(i,sse))
    inertias.append(sse)
```

```python
plt.plot(range(2,20,2),inertias)
plt.show
```

```python
def k_means_inertia_calculation(datasets: np.ndarray, k: int):
    """ Calculate the inertia of K-means algorithm

    Arg:
        datasets(np.ndarray): The dataset that will be used to calculate the inertia.
        k(int): The number of clusters.

    Return:
        The inertia of K-means algorithm.
    """
    pass

    # Code start here
    kmeans = KMeans(n_clusters=k, random_state=0).fit(datasets)
    return kmeans.inertia_
```

```python
# Please replace the path if needed
fileName = "living_room.jpg"
img = io.imread(fileName)
img = img.reshape(img.shape[0] * img.shape[1], 3)
inertia = []
for k in range(2, 20, 2):
    SSE = k_means_inertia_calculation(img, k)
    inertia.append(SSE)
    print("k=%s\t SSE = %f" % (k, SSE))
plt.plot(range(2, 20, 2), inertia)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.title("Inertia of K-means algorithm")
plt.show()
```
