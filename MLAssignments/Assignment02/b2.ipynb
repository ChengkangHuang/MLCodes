```python
import numpy as np    # for numerical opertations
import pandas as pd   # for data analysis and manipulation

# for creating the classification model by building a decision tree
from sklearn.tree import DecisionTreeClassifier   

# for splitting data arrays into two subsets: for training data and for testing data
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('tripadvisor_review.csv')
```

```python
df.head()
```

```python
X = df.drop(['UserID'], axis=1)
X.head()
```

```python
def setLabel(score: float) -> str:
    """ Check the score from the data and give back the label

    Base on the rating score, compare with the enum "Rating" to set
    the label for each record.

    Arg:
        score(float): The number from each record.

    Return:
        resultStr(str): The label base on different input.

    """
    pass

    resultStr = ""
    if 0.0 <= score <= 1.0:
        resultStr = "Terrible"
    elif 1.0 < score <= 2.0:
        resultStr = "Poor"
    elif 2.0 < score <= 3.0:
        resultStr = "Average"
    elif 3.0 < score <= 4.0:
        resultStr = "VeryGood"
    elif 4.0 < score <= 5.0:
        resultStr = "Excellent"
    else:
        resultStr = "No rating scores"
    return resultStr

labels = (df['Art_galleries'] + df['Dance_clubs'] + df['Juice_bars'] + df['Restaurants'] + df['Museums'] + df['Resorts'] + df['Picnic_spots'] + df['Beaches'] + df['Theaters'] + df['Religious_institutions']) / 10
y = labels.apply(setLabel)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)
```

```python
y_pred = dt_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
def dtcRun1(X, y, c="gini", s="best", md=7, mln=5):
    # Randomly split 75% data for training, then the rest 25% data for testing
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3)
    dt_clf = DecisionTreeClassifier(criterion=c, splitter=s, max_depth=md, max_leaf_nodes=mln)
    dt_clf.fit(X_train, y_train)
    y_pred = dt_clf.predict(X_test)
    return accuracy_score(y_test, y_pred)
runResult = dtcRun1(X, y, c="gini")
print("Criterion = gini", runResult)
```

```python
def dtcRun(X, y, c="gini", s="best", md=7, mln=5):
    result = []
    # 5 runs
    for i in range(5):
        accuracyScore = 0
        # Run 50 times
        for z in range(50):
            # Randomly split 75% data for training, then the rest 25% data for testing
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3)
            dt_clf = DecisionTreeClassifier(criterion=c, splitter=s, max_depth=md, max_leaf_nodes=mln)
            dt_clf.fit(X_train, y_train)
            y_pred = dt_clf.predict(X_test)
            accuracyScore += accuracy_score(y_test, y_pred)
        result.append(accuracyScore / 50)
    return np.mean(np.array(result)), result

mean, runResult = dtcRun(X, y, c="gini")
print("Criterion = gini", mean, runResult)
```
