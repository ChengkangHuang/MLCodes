```python
import csv
countrysize = {}

with open ("cities2.csv") as file:
    reader =  csv.DictReader(file)
    for row in reader:
        try:
            countrysize[row["CountryCode"]] += int(row["Population"])
        except:
            countrysize[row["CountryCode"]] = int(row["Population"])

for country in sorted(countrysize.items()):
    print(country[0], country[1])
```

```python
import numpy as np
from matplotlib import pyplot as plt

randoms = np.random.randint(0, 100, (1000, 2))
plt.scatter(randoms[:, 0], randoms[:, 1], marker="v", c="aqua")
plt.title("Aqua Triangles")
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

randoms1 = (np.random.random((1000, 2)) - 0.5 * np.random.random((1000, 2)) - 0.5)
randoms2 = (np.random.random((1000, 2)) - 0.5 * np.random.random((1000, 2)) - 0.5)

plt.scatter(randoms1[:, 0], randoms1[:, 1], marker=".", c="purple")
plt.scatter(randoms2[:, 0], randoms2[:, 1], marker=".", c="pink")

plt.title("Pink vs Purple")
plt.show()
```
