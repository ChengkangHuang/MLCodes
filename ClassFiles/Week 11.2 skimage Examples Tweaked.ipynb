```python
# skimage stock image
# https://scikit-image.org/docs/stable/api/skimage.io.html
cat = data.chelsea()
plt.figure()
plt.title("data.chelsea()")
plt.imshow(cat)
plt.axis('off')
plt.show()
```

```python
# make cat grayscale using camap
cat = data.chelsea()
plt.imshow(cat[:,:,1], cmap='gray')
plt.axis('off')
plt.show()
```

```python
# make cat grayscale using camap
cat = data.chelsea()
plt.imshow(cat[:,:,0], cmap='gray')
plt.axis('off')
plt.show()
```

```python
# make cat grayscale using camap
cat = data.chelsea()
plt.imshow(cat[:,:,2], cmap='gray')
plt.axis('off')
plt.show()
```

```python
# make cat grayscale first try
intensities = cat[:,:].mean(axis=2).round()
for row in range(cat.shape[0]):
    for col in range(cat.shape[1]):
        cat[row,col]=intensities[row,col]*3   # average of rgb values

plt.figure()
plt.title("data.chelsea() - grayscale")
plt.imshow(cat)
plt.axis('off')
plt.show()
```

```python
# make cat grayscale second try
from skimage.color import rgb2hsv
from skimage.color import hsv2rgb
cat = data.chelsea()
cat = rgb2hsv(cat)   # convert to hsv
cat[:,:,1] = 0       # change saturation to 0
cat = hsv2rgb(cat)   # convert back to rgb

plt.figure()
plt.title("data.chelsea() - grayscale second try")
plt.imshow(cat)
plt.axis('off')
plt.show()
```

```python
# save grayscale cat
io.imsave("cat.png", cat)
```
