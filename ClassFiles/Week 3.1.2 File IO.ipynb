```python
def readFile(fileName):
    with open(fileName) as file:
        for line in file:
            for word in line.split():
                print(word)

readFile("fileToRead.txt")
```

```python
def writeFile(fileName, contents):
    with open(fileName, "w") as file:
        for line in contents:
            file.write(line+"\n")

readFile("fileToWrite.txt")
writeFile("fileToWrite.txt", ["this is line 1", "this is line 2"])
```

```python
def writeFile(fileName, contents):
    with open(fileName, "a") as file:
        for line in contents:
            file.write(line+"\n")

readFile("fileToWrite.txt")
writeFile("fileToWrite.txt", ["this is line 1", "this is line 2"])
```

```python
def readFile(fileName):
    with open(fileName, "r", encoding="latin-1") as file:
        for line in file:
            for word in line.split():
                print(word)

readFile("spamfile.txt")
```

```python
def count_spam_word(word):
    count = 0
    with open("spamfile.txt", encoding="latin-1") as file:
        for line in file:
            if word in line.lower().split():
                count+=1
    return count

wordCount = count_spam_word("best")
print("The word best appear " + str(wordCount) + " times")
```
