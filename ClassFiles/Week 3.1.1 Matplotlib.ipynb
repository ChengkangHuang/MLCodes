```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1, 11)
y = x * 2
```

```python
plt.plot(x, y, color="green")
plt.title("X vs Y")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.show()
```


#### Scatter PLot

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
a = [8, 1, 7, 4, 5, 6, 7, 8, 9]
b = [1, 7, 8, 9, 1, 4, 5, 3, 8]
```

```python
plt.scatter(x, a, marker = "*", c = "g", s = 50)
plt.scatter(x, b, marker = "o", c = "r", s = 50)
plt.show()
```

```python
plt.subplot(1, 3, 1)    # has 1 row, 3 cols and this is the 1st col
plt.scatter(x, a, marker="*", c="g",s=50)

plt.subplot(1, 3, 2)    # has 1 row, 3 cols and this is the 2nd col
plt.scatter(x, b, marker="o", c="r",s=50)
```

### Bar PLot

```python
student = {"rami" : 87, "mat" : 56, "noor" : 27}
student
```

```python
names = list(student.keys())    # type casting to a list
ages = list(student.values())
```

```python
names
```

```python
ages
```

```python
plt.bar(names, ages)
plt.title("Bar plots")
plt.xlabel("Names")
plt.ylabel("Ages")
plt.grid(True)
```

```python
plt.barh(names, ages, color = "y")
plt.title("Bar plots")
plt.xlabel("Names")
plt.ylabel("Ages")
plt.grid(True)
```

```python
plt.figure(1)   # Create a new figure, or activate an existing one
data = np.array([[5, 4], [2, 6], [9, 9], [3, 7]])
data
```

```python
data = np.random.rand(100, 2) * np.random.randint(1, 100) + np.random.randint(1, 100)   # rand(rows, cols)
data
```
