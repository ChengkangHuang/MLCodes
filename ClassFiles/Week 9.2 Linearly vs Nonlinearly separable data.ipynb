# Linearly separable data 



```python
# Perceptron is a linear classifier. It creates a line to separate the classes

# Equation of the line is y = mx + b
# b = the intercept of the line
# m = slope of the line

# Activation of the perceptron is:
#     w1x1 + w2x2 > T   replace the x1 with x and x2 with y
#     w1x + w2y > T

# The decision boundary is 
#     w1x + w2y = T
#     w2y = -w1x + T   divide the eqation by y

#     y = (-w1/w2)x + T/w2
#     y = mx + b
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# data=pd.read_csv('data1.csv')  # Linearly separable data
# data=pd.read_csv('data2.csv')  # Non_linearly separable data
# data=pd.read_csv('data3.csv')  # Linearly separable data
data=pd.read_csv('data4.csv')    # Linearly separable data
```

```python
data
```

```python
label=data['Label']
plt.scatter(data['F1'],data['F2'],c=label)  
```

```python
plt.scatter(data['F1'],data['F5'],c=label)   
```

```python
X = data.drop(['Label'], axis=1)
y = data['Label']
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=4)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
from sklearn.linear_model import Perceptron
p_clf = Perceptron(verbose=True)
p_clf
```

```python
p_clf.fit(X_train, y_train)
```

```python
y_pred = p_clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
