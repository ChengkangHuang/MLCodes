```python
import pandas as pd
iris1 = pd.read_csv('iris1.csv')
iris2 = pd.read_csv('iris2.csv')
```

```python
iris1.head()
```

```python
iris2.head()
```

```python
# NaN means not a number, null value

# iris1.dropna()      # Remove all rows with NULL values from the Dataframe
# iris1.fillna(0)     # Replace NULL values with the number 0

# Not only use for whole dataset, but also use for signle column
iris1['sepal.length'].fillna(iris1['sepal.length'].mean(), inplace = True)
iris1['petal.length'].fillna(iris1['petal.length'].mean(), inplace = True)
iris1
```

```python
df = pd.concat([iris1, iris2], axis=0)      # Row wise concatenation
# df = pd.concat([iris1, iris2], axis=1)      # Column wise concatenation 
df
```

```python
df['variety'].unique()
```

```python
df[df['variety'] == 'Setosa']
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode target labels with value between 0 and n_classes-1
label_encoder = LabelEncoder()
label_encoder
```

```python
df['variety'] = label_encoder.fit_transform(df['variety'])      # Fit label encoder and return encoded labels
df['variety']
```

```python
df.head()
```

```python
# df['vareity'].map({'Setosa' : 1, "Virginica" : 2, 'Versicolor' : 3})
```

```python
# Seperating features from labels
X = df.drop(columns='variety')      # (Column) X = df.drop(columns='variety', axis = 1)
# X = df.drop([2, 3, 4], axis=0)      # (Row) X = df.drop([2, 3, 4], axis = 0)
X
```

```python
y = df['variety']
y
```

```python
# Spliting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
X_train
```

```python
# Create GaussianNB Object
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model
```

```python
# Train the classifier with the training sets
nb_model.fit(X_train, y_train)      # Fit Gaussian Naive Bayes according to X, y
```

```python
# Calulate the mean accuracy
nb_model.score(X_train, y_train)
```

```python
nb_model.score(X_test, y_test)
```

### Result

- The model is overfitting for test_size = 0.3, random_state = 20 (accuracy score for the train set 0.980) > (accuracy score for test set 0.889)

- (The Best Result) The model is overfitting for test_size = 0.4, random_state = 3 (accuracy score for the train set 0.944) > (accuracy score for test set 0.983)

- The model is overfitting for test_size = 0.2, random_state = 10 (accuracy score for the train set 0.950) > (accuracy score for test set 0.966)
