```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data = load_iris()
data
```

```python
feature_set=data.data
feature_set
```

```python
label_set = data.target
label_set       # Original labels
```

```python
from sklearn.cluster import KMeans
# km_clustering = KMeans(n_clusters=3)  # inertia = 78.85
# km_clustering = KMeans(n_clusters=3, verbose=1)   # inertia = 78.85
km_clustering = KMeans(n_clusters=4, verbose=1)     # inertia = 57.228
# km_clustering = KMeans(n_clusters=5, verbose=1)     # inertia = 46.446
cls_model = km_clustering.fit(feature_set)
cls_model
```

```python
cls_pred_labels = cls_model.labels_     # Clustered labels
cls_pred_labels
```

```python
centroids = cls_model.cluster_centers_
centroids
# ndarray if shape (n_clusters, n_features)
# array(1st [cluster centers for feature 1, cluster centers for feature 2])
#      (2nd [cluster centers for feature 1, cluster centers for feature 2])
```
